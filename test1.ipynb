```python
#original poulation_gender.csv파일을 population_gender_202510.csv파일로 변경하는 코드
#쓸모없는 열 날리고 정리함

import pandas as pd
df=pd.read_csv("./original/gender_population.csv",encoding="cp949")
df=df.drop(['2025년10월_계_연령구간인구수','2025년10월_남_총인구수','2025년10월_여_총인구수'],axis=1)
df
df.rename(columns={"행정구역":"area",
                   '2025년10월_계_총인구수':'total',
                   "2025년10월_계_0~9세": "10세미만", "2025년10월_계_10~19세": "10대","2025년10월_계_20~29세": "20대","2025년10월_계_30~39세": "30대","2025년10월_계_40~49세": "40대","2025년10월_계_50~59세": "50대","2025년10월_계_60~69세": "60대","2025년10월_계_70~79세": "70대","2025년10월_계_80~89세": "80대","2025년10월_계_90~99세": "90대","2025년10월_계_100세 이상": "100세이상",

                   '2025년10월_남_연령구간인구수':'total',
                   "2025년10월_남_0~9세": "10세미만", "2025년10월_남_10~19세": "10대","2025년10월_남_20~29세": "20대","2025년10월_남_30~39세": "30대","2025년10월_남_40~49세": "40대","2025년10월_남_50~59세": "50대","2025년10월_남_60~69세": "60대","2025년10월_남_70~79세": "70대","2025년10월_남_80~89세": "80대","2025년10월_남_90~99세": "90대","2025년10월_남_100세 이상": "100세이상",
                   
                   '2025년10월_여_연령구간인구수':'total',
                   "2025년10월_여_0~9세": "10세미만", "2025년10월_여_10~19세": "10대","2025년10월_여_20~29세": "20대","2025년10월_여_30~39세": "30대","2025년10월_여_40~49세": "40대","2025년10월_여_50~59세": "50대","2025년10월_여_60~69세": "60대","2025년10월_여_70~79세": "70대","2025년10월_여_80~89세": "80대","2025년10월_여_90~99세": "90대","2025년10월_여_100세 이상": "100세이상",}, inplace=True)
df["area"] = df["area"].str.replace(r"\s*\(.*\)", "", regex=True)

df["area"] = df["area"].replace({
    "서울특별시": "서울",
    "부산광역시": "부산",
    "대구광역시": "대구",
    "인천광역시": "인천",
    "광주광역시": "광주",
    "대전광역시": "대전",
    "울산광역시": "울산",
    "세종특별자치시": "세종",
    "경기도": "경기",
    "강원특별자치도": "강원",
    "충청북도": "충북",
    "충청남도": "충남",
    "전북특별자치도": "전북",
    "전라남도": "전남",
    "경상북도": "경북",
    "경상남도": "경남",
    "제주특별자치도": "제주",
    "전국": "한국"
})

# 전체, 남성, 여성 데이터 분리
df_gender = df.iloc[:, 0:13].copy()
df_men    = df.iloc[:, [0] + list(range(13, 25))].copy()
df_women  = df.iloc[:, [0] + list(range(25, 37))].copy()

# gender 열을 2번째 열로 삽입
df_gender.insert(1, "gender", "전체")
df_men.insert(1, "gender", "남성")
df_women.insert(1, "gender", "여성")

# 열 이름 통일: 지역 + gender + 연령대
age_columns = df.columns[1:13].tolist()
new_columns = ["area", "gender"] + age_columns

df_gender.columns = new_columns
df_men.columns    = new_columns
df_women.columns  = new_columns

# 세 데이터프레임 합치기
df_combined = pd.concat([df_gender, df_men, df_women], ignore_index=True)

# CSV 저장
df_combined.to_csv("./csv/gender_population_202510.csv", index=False, encoding="cp949")
df_combined

#남자만 필터
# df_male_only = df_combined[df_combined["gender"] == "남성"]
# df_male_only
```
